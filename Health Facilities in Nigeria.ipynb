```python
facility_count =(
     df.groupby('Facility Type')['Facility Id']
     .count()
     .reset_index(name='Total Facilities'))

facility_count =facility_count.sort_values(by='Total Facilities', ascending=False)

plt.figure(figsize=(10,6))
bars = plt.barh(
    facility_count['Facility Type'],
    facility_count['Total Facilities'],
    color = '#003366')

plt.ylabel('Facility Type', fontsize = 14)
plt.xlabel('Number of Facilities', fontsize = 14)
plt.title('Breakdown of Total Healthcare Facilities across Facility Type',fontweight='bold', fontsize=15)

for index, value in enumerate(facility_count['Total Facilities']):
    plt.text(value +1, index,str(value), va ='center')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
service_cols =[
    'Antenatal Care',
    'Child Measles Immunization',
    'Maternal Health Delivery Services',
    'Family Planning',
    'Skilled Birth Attendant',
    'C_Section'
]
poor_services = df.groupby('Facility Type')[service_cols].apply(lambda x:(~x).sum())
facility_count= df.groupby('Facility Type').size()
poor_service_percent = (poor_services.div(facility_count,axis =0).mean(axis =1)*100).round(1)
poor_service_percent = poor_service_percent.reset_index()
poor_service_percent.columns=['Facility Type', 'Poor Services%']
poor_service_percent =poor_service_percent.sort_values(by = 'Poor Services%', ascending=False)

plt.figure(figsize=(10,6))
bars = plt.barh(
    poor_service_percent['Facility Type'],
    poor_service_percent['Poor Services%'],
    color = 'red')
plt.ylabel('Facility Type', fontsize = 14)
plt.xlabel('Poor Maternal and Child Healthcare Service (%)', fontsize = 14)
plt.title('Gaps in Maternal and Child Healthcare Delivery(%) by Facility Type',fontweight='bold', fontsize=14)
for index, value in enumerate(poor_service_percent['Poor Services%']):
    plt.text(value +1, index,str(value), va ='center')
plt.tight_layout()
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
